## 01 Determine whether Grover adaptive search is successful
### Description
In Quantum algorithm, Grover adaptive search is a widely accepted method to Grover adaptive search find the minimum of an optimization problem. As shown in the following figure, in each iteration of the algorithm, there is a threshold "y" to determine the succeed or fail of the algorithm execution.

image 01

Thus, after each iteration of Grover adaptive search, we need to validate whether the solution is succeed or not. In particular, if the output value "u" is smaller than the threshold "y", Grover adaptive search will return "succeed".
### Input
Input two integers to denote the function value "u" and the threshold "y".
### Output
Output 'Succeed' or 'Fail'.

**Sample Input 1** `12 16`

**Sample Input 2** `100 34`

## 02 Application of greatest common divisor
### Description
A rectangular piece of cardboard has length "**a**" cm and width "**b**" cm. Now you want to cut it into a number of small squares, and you want the sides of the square to be a whole number of centimeters long, so what is the maximum side length of the square?



To solve this problem, you can use the method of "greatest common divisor". To make the largest area of the small square, which means to make the largest side length, this is the problem of finding the "greatest common divisor" of "**a**" and "**b**".

HINTs: search greatest common divisor in the math package
### Input
Input two integers to denote the length and width of the rectangular (i.e., "**a**" and "**b**").
### Output
The greatest common divisor of "a" and "b".

**Sample Input 1** `4 6`

**Sample Output 1** `2`

**Sample Input 2** `9 15`

**Sample Output 2** `3`
### Hint
Enumerate all the numbers no larger than "**a**" and "**b**" with for-loop.

Determine whether the current number can divide both "**a**" and "**b**".

````python
a,b = map(int,input().split())
def gcd(a,b):
    if b == 0:
        return a
    else:
        return gcd(b,a%b)
print(gcd(a,b))
````

## 03 Print hollow rectangle
### Description
You need to print an m * n hollow rectangle as shown in the following figure, which is an example of 8 * 8 hollow rectangle.


### Input
Input two integers to denote the length and width of the rectangular (i.e. m and n)
### Output
Print the "m * n"hollow rectangle.

**Sample Input 1** `7 8`

**Sample Output 1**
```
********

*      *
*      *
*      *
*      *
********
```
**Sample Input 2** `5 7`

**Sample Output 2**
```
*******

*     *
*******
```
### Hint
Analyze the shape layer by layer, print ' * ' for the boundary and print ' '(space) to fill inside, to find the relationship between the number of cycles and the printed graphics.

Use "**for-loop**" and "**if-else**" to print the graph.

````python
a,b = map(int,input().split())
for i in range(1,a+1):
    for j in range(1,b+1):
        if i==1 or i==a or j==1 or j==b:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

## 06 Multiplication table
### Description
Multiplication table is an interesting task to check your basic loop programming ability.

An integer size, 0<size<10.
### Output
A lower triangle of the multiplication table.

There's a space at the end of each line.

**Sample Input 1** `5`

**Sample Output 1**
```
1x1=1 
2x1=2 2x2=4 
3x1=3 3x2=6 3x3=9 
4x1=4 4x2=8 4x3=12 4x4=16
```

## 07 List Overlap
### Description
Given two lists, your task is to return a new list that contains the "intersection elements"(co-appear in both lists, without duplication). Make sure your program works on two lists of different sizes.
### Input
two list of int a,b

the length of a and b may not be equal
### Output
a list of non-repeat common elements in ascending order

**Sample Input 1**
```
1 2 3 4 5 3 88
6 4 5 5 5 5 3 1 9 10 456 23
```
**Sample Output 1** `1 3 4 5`

**Sample Input 2**
```
13 12 15 3 12 13 7 8 18 6
4 15 15 8 0 3 17 8 1 15
```
**Sample Output 2** `3 8 15`

**Sample Input 3**
```
15 6 12 16 9 13 3 15 17 17
15 8 13 9 12 9 11 2 10 5
```
**Sample Output 3** `9 12 13 15`
### Hint
Brute force: two loops O(n^2)

Pruning brute force: sort before loops O(nlogn)

set: O(n)

You can solve this problem with numpy

````python
import numpy as np
A = np.array([int(x) for x in input().split()])
B = np.array([int(x) for x in input().split()])
inter = np.intersect1d(A, B)
print(' '.join((str(i) for i in inter)))
````

## 08 Convert two lists intodictionary
### Description
Given a list of keys and a list of values, your need 'convert' them into a dictionary.

Your task is to implement to '**makedict**' function, two list(keys and values) is passed to the function, you need to return the dictionary object.

All the key and value are in str type.

**Hints**

You could first try to do this task by yourself and then find a method zip() from 'dictionary' to make it done.
### Input
two list separated by a semicolon ';'  in one line
### Output
Return the resulting dictionary in the function

**Sample Input 1** `ten twenty thirty;10 20 30`

**Sample Output 1** `{'ten': '10', 'twenty': '20', 'thirty': '30'}`

**Sample Input 2** `a b c d;1 2 3 4`

**Sample Output 2** `{'a': '1', 'b': '2', 'c': '3', 'd': '4'}`

````python
def makedict(keys,values)-> dict:
  a_dict = dict(zip(keys,values))
  #write your code here
  return a_dict


_input = input()
keys,values = _input.split(';')
keys = keys.split()
values = values.split()
print(makedict(keys,values))
````
